### Notebook 1:  Clustering con K-means
En este notebook vamos a implementar el algoritmo de clustering K-means en R. El objetivo es entender como funciona este algoritmo y  aprender a utilizarlo con un ejemplo simple. Se incluye además la estrategia para definir el número de grupos k. Al final se propone un ejercicio con nuevos datos, con el fin de aprender a utilizar el algoritmo. 

```r
library(cluster)
library(ggplot2)

#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("milk1.dat",row.names=1,dec=".",header=TRUE)
print(datos,header=TRUE)
summary(datos)
boxplot(datos)
```

**Ejercicio 1:** describa los datos y discuta la pertinencia de realizar un análisis de clustering o agrupamiento

# Respuesta
Los datos corresponden a las concentraciones de agua, proteína, grasa, lactosa y calcio que contienen la leche que producen ciertos mamiferos. Realizar un agrupamiento de ésta información, nos ayudará a visualizar mejor los datos ya que a simple vista es un poco tedioso ver tanta información, también nos ayuda ya que al ser muchas variables explicativas es dificil ordenar todos los datos y un agrupamiento nos beneficiará en ese aspecto. También nos ayuda ver ciertas similitudes de la leche que producen ciertos mamiferos, ya que los que están en un mismo grupo, producen una leche con concentraciones similares.

```r
#ingresar valor de k0 
k0=3
cl<-kmeans(datos,k0,iter.max=1000,nstart=10)
#grafica los datos coloreados con el grupo asignado en planos de pares de variables
pairs(datos,col=cl$cluster)
```

**Ejercicio 2:** Interprete y comente los resultados obtenidos. 

# Respuesta
El gráfico nos ayuda a ver la distribución de los grupos haciendo una relación entre dos de las vriables, cada grupo tiene un color, rojo, negro y verde. Por ejemplo entre el agua y la protenia, se ve que los que poseen un alto nivel de agua y no tanto de proteina están en el grupo rojo, los que poseen un valor alto de proteina y regular de agua estan en el grupo negro y finalmente los que poseen un bajo nivel de agua y alto de proteina están de color verde. Así cada grafico indica el grupo de los datos segun el nivel de las variables.

```r
#Análisis de Componentes Principales para graficar los grupos obtenidos
#Calcula plano de componentes principales y grafica las observaciones con el número de su grupo
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,cl$centers)
res2
cl
biplot(res,xlabs=cl$cluster)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:k0,pch=19)
```

```r
#Estrategia para elegir K, el numero de grupos
#J suma cuadrado de las distancias por cada k utilizado
#K-means
J<-rep(0,20)
fin_iter<-20


for (k in 1:fin_iter){
     cl<-kmeans(datos,k,iter.max=1000,nstart=10)
	 J[k]<-cl$tot.withinss
}


plot(1:fin_iter,J,xlab="k",ylab="J",type="p",main="J v/s k en k-means",col=2) 
```

**Ejercicio 4:** de acuerdo al gráfico obtenido, cuál es el valor de K mas apropiado?
Agregue código para automatizar este criterio.

# Respuesta
El valor mas apropiado es un k=3 ya que se ve en el gráfico que es el punto de inflexión de la curva, también se ve que la diferencia de la variable J que indica la función de perdida empieza a decaer con un k superior a 3

```r
J<-rep(0,20)
fin_iter<-20
d1 = 0
d2=0
tasa=0
tasa_maxima=0
k1=0
for (k in 1:fin_iter){
    cl<-kmeans(datos,k,iter.max=1000,nstart=10)
    J[k]<-cl$tot.withinss
    if(k!=1 && k!=20){
        d1= abs(J[k]-J[k-1])
        d2 = J[k]-J[k+1]
        tasa = d1/d2
    }
    if(tasa > tasa_maxima ){
        tasa_maxima = tasa
         k1 = k+1 #YA QUE SE EMPIEZA A ITERAR DESDE EL K=1
    }
}


plot(1:fin_iter,J,xlab="k",ylab="J",type="p",main="J v/s k en k-means",col=2)

print("Calculado la tasa de las diferencias de las distancias entre los k nos da un k óptimo:")
print(k1)

```

# Calculando un k óptimo
